```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open('pop_coding_3.4.pickle', 'rb') as f:
    data = pickle.load(f)

with open('tuning_3.4.pickle', 'rb') as f:
    tuning_data = pickle.load(f)
```

Now I plot the tuning curves and save the firing rates and rmax for each neuron.

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
stim = tuning_data['stim']

firing_rates = []
rmax = []

for i in range(4):
    row = i // 2
    col = i % 2
    mean_firing_rate = np.mean(tuning_data['neuron' + str(i+1)], axis=0) / 10  # 10 seconds
    rmax_i = np.max(mean_firing_rate)
    
    firing_rates.append(mean_firing_rate)
    rmax.append(rmax_i)
    
    ax[row][col].plot(stim, mean_firing_rate);
    ax[row][col].set_title('Neuron ' + str(i+1))
```
