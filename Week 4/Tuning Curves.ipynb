```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('tuning_3.4.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
data.keys()
```

## Raster plot for neuron 1

```python
plt.eventplot(data['neuron1']);
```

```python
firing_rates = np.mean(data['neuron1'], axis=0)
```

```python
firing_rates.shape
```

## Tuning curve for neuron 1(?)

```python
plt.plot(data['stim'], firing_rates);
```

## Firing rates for all neurons

```python
firing_rate1 = np.mean(data['neuron1'], axis=0)
firing_rate2 = np.mean(data['neuron2'], axis=0)
firing_rate3 = np.mean(data['neuron3'], axis=0)
firing_rate4 = np.mean(data['neuron4'], axis=0)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(6, 6))
ax[0][0].hist(firing_rate1);
ax[0][1].hist(firing_rate2);
ax[1][0].hist(firing_rate3);
ax[1][1].hist(firing_rate4);
```

```python
fano1 = np.var(firing_rate1) / np.mean(firing_rate1)
fano2 = np.var(firing_rate2) / np.mean(firing_rate2)
fano3 = np.var(firing_rate3) / np.mean(firing_rate3)
fano4 = np.var(firing_rate4) / np.mean(firing_rate4)
print(fano1, fano2, fano3, fano4)
```

## Tuning curves for all neurons

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
ax[0][0].plot(data['stim'], firing_rate1);
ax[0][1].plot(data['stim'], firing_rate2);
ax[1][0].plot(data['stim'], firing_rate3);
ax[1][1].plot(data['stim'], firing_rate4);
```
